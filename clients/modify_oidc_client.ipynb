```python
bearer_token = clb_oauth.get_token()
```

```python
import requests
import json

# the API to the OIDC client app
API_PROD = 'https://wiki.ebrains.eu/rest/v1/oidc/clients'
API_INT  = 'https://wiki-int.ebrains.eu/rest/v1/oidc/clients'
API = API_PROD

# the clientId we use in this notebook
clientId = "restricted-access-email"
```

```python
response = requests.get(
    f'{API}/{clientId}',
    headers={'Authorization': 'Bearer %s' % bearer_token}
)

if response.status_code == requests.codes.ok:
    ebrainsClientResponse = response.json()
    print(json.dumps(ebrainsClientResponse, indent=4))
else:
    print("FAILED", response.content)
```

```python
clientId_wizard = "ebrains-wizard-dev"
```

```python
response = requests.get(
    f'{API}/{clientId_wizard}',
    headers={'Authorization': 'Bearer %s' % bearer_token}
)

if response.status_code == requests.codes.ok:
    ebrainsClientResponse_wizard = response.json()
    print(json.dumps(ebrainsClientResponse_wizard, indent=4))
else:
    print("FAILED", response.content)
```

```python
# Change the description of the OIDC client
ebrainsClientResponse['client']['standardFlowEnabled'] = True;
   
response = requests.put(
    f'{API}/{clientId}',
    headers={'Authorization': 'Bearer %s' % bearer_token},
    json=ebrainsClientResponse
)

if response.status_code == requests.codes.ok:
    ebrainsClientResponse = response.json()
    print(json.dumps(ebrainsClientResponse, indent=4))
else:
    print("FAILED", response.content)
```
